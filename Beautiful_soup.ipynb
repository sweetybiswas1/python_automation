```python
import requests
from bs4 import BeautifulSoup
```

```python
url='https://codewithharry.com'
```

```python
# S1: Get the HTML

req = requests.get(url)
htmlcontent=req.content
print(htmlcontent)
```

```python
# S2: Parse the HTML

soup=BeautifulSoup(htmlcontent,'html.parser')
print(soup)
```

```python
print(soup.prettify)
```

```python
# S3: HTML tree traversal

title=soup.title # Get the title of the HTML page

print(title)
```

```python
# types of object 
print(type(title)) # 1.Tag

print(title.string)
print(type(title.string)) # 2. NavigableString

print(type(soup)) # 3. BeautifulSoup

# 4. Comment
x="<p><!-- this is a comment --></p>"
soup2=BeautifulSoup(x)
print(type(soup2.p.string))
```

```python
# Get all the paragraphs from the page
paras=soup.find_all('p')
print(paras)
```

```python
# Get all the anchor tags from the page
anchors=soup.find_all('a')
print(anchors)
```

```python
print(soup.find('p')) #get first element in the html page
```

```python
#get classes of any element in the html page
print(soup.find('p')['class']) 
```

```python
# Find all the elements with class lead

soup.find_all('p',class_='lead')
```

```python
#get the text from the tags/soup
print(soup.find('p').get_text()) 
```

```python
#get all text of any html page
print(soup.get_text()) 
```

```python
#Get all the links on the page
anchors=soup.find_all('a')
for link in anchors:
    print(link.get('href'))
```

```python
#Get all the links on the page
anchors=soup.find_all('a')
all_links=set()
for link in anchors:
    if link.get('href') != '#':
        links='https://codewithharry.com' +link.get('href')
        all_links.add(links)
print(all_links)
```

```python
url='https://www.flipkart.com/search?q=mobiles&sid=tyy%2C4io&as=on&as-show=on&otracker=AS_QueryStore_OrganicAutoSuggest_1_4_na_na_na&otracker1=AS_QueryStore_OrganicAutoSuggest_1_4_na_na_na&as-pos=1&as-type=RECENT&suggestionId=mobiles%7CMobiles&requestId=853810f0-df1e-4c7d-84fd-75661688e436&as-searchtext=mobi'
import requests
from bs4 import BeautifulSoup
req = requests.get(url)
htmlcontent=req.content
soup = BeautifulSoup(htmlcontent,'html.parser')
print(soup)
```

```python
product=[]
price=[]
```

```python
x=soup.find('div',attrs={'class':'_4rR01T'})
x.text
```

```python
for i in soup.findAll('a',href=True,attrs={'class':'_1fQZEK'}):
    name=i.find('div',attrs={'class':'_4rR01T'})
    pric=i.find('div',attrs={'class':'_30jeq3'})
    product.append(name.text)
    price.append(pric.text)
    
    
print(product)
print(price)
    
```

```python
import pandas as pd

df=pd.DataFrame({'product_name':product,'Prices':price})
```
